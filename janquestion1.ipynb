```python
import pandas as pd
import sqlite3
df=pd.read_excel(r'E:\programming\Data Science Course\Group project files\flatten_data_table-main\wikipedia_dataset_flat.xlsx',index_col=[0])
df= df.iloc[2:] 
```

```python
df
```

```python
jandata=df.query('date=="2016-01-01"')
jandata
```

```python
jandata[["title","last wikipedia"]]=jandata['page'].str.split(".",n=1,expand=True)
jandata
```

```python
#as_index to set the column names 
split=jandata.groupby(['search title'],as_index=False)['visits'].sum()
split

```

```python
conn=sqlite3.connect('Trending searches Jan.db')
```

```python
c=conn.cursor()
```

```python
reate_table="CREATE TABLE search(searchtitleTEXT,visits INT)"
```

```python
c.execute(reate_table)
conn.commit()
```

```python
split.to_sql('search',conn,if_exists='append',index=False)
```
